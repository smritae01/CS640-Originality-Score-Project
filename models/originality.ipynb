```python
import pandas as pd

weights_df = pd.read_csv('../data/normalized_weights.csv')

originality_df = pd.read_csv('../data/originality_score.csv')
```

```python
weights_df
```

```python
openai_weight = weights_df["Weight"][0]
gptzero_perplexity = weights_df["Weight"][1]
gptzero_burstiness = weights_df["Weight"][2]
detectgpt_zscore = weights_df["Weight"][3]
gptzero_weight = weights_df["Weight"][4]
detectgpt_weight = weights_df["Weight"][5]
```

```python
weights = {
    'Open AI Classfier': openai_weight,
    'GPT Zero (perplexity)': gptzero_perplexity,
    'GPT Zero (burstiness)': gptzero_burstiness,
    'DetectGPT (z score)': detectgpt_zscore,
    'GPT Zero ': gptzero_weight,
    'DetectGPT': detectgpt_weight
}
```

```python
originality_df
```

```python
def scale_column(col):
    return (col - col.min()) / (col.max() - col.min()) * 100

originality_df['Open AI Classfier'] = scale_column(originality_df['Open AI Classfier'])
originality_df['GPT Zero (perplexity)'] = scale_column(originality_df['GPT Zero (perplexity)'])
originality_df['GPT Zero (burstiness)'] = scale_column(originality_df['GPT Zero (burstiness)'])
originality_df['DetectGPT (z score)'] = scale_column(originality_df['DetectGPT (z score)'])
originality_df['GPT Zero '] = scale_column(originality_df['GPT Zero '])
originality_df['DetectGPT'] = scale_column(originality_df['DetectGPT'])
```

```python
originality_df['originality_score'] = originality_df.apply(lambda row: sum(row[feature] * weight for feature, weight in weights.items()), axis=1)
```

```python
originality_score = originality_df[list(weights.keys())].mul(list(weights.values())).sum(axis=1)
```

```python
originality_df['originality_score'] = (originality_df['originality_score'] - originality_df['originality_score'].min()) / (originality_df['originality_score'].max() - originality_df['originality_score'].min()) * 100
```
